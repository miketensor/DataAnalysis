```python
import numpy as np
import matplotlib.pyplot as plt 

rng = np.random.RandomState(42)
x = 10 * rng.rand(200)

def model(x, sigma=0.3):
    fast_oscillation = np.sin(5 * x)
    slow_oscillation = np.sin(0.5 * x)
    noise = sigma * rng.randn(len(x))

    return slow_oscillation + fast_oscillation + noise

y = model(x)
plt.errorbar(x, y, 0.3, fmt='o')
```

```python
import numpy as np
import matplotlib.pyplot as plt
import random
import matplotlib.patches as mpatches

rgn = np.random.RandomState(53)


def model(x):
    return np.sin(5 * x) + np.sin(0.5 * x) + 0.3 * rgn.rand(len(x))


X = 10 * rgn.rand(50)

Y = model(X)
plt.errorbar(X, Y, 0.3, fmt='o')

from sklearn.ensemble import RandomForestRegressor

forest = RandomForestRegressor(500)
forest.fit(X.reshape(-1,1),Y)

XPredict = np.linspace(0,10,1000)
YPredict = forest.predict(XPredict.reshape(-1,1))
YTrue = model(XPredict)

red_patch = mpatches.Patch(color='red', label='The true data')
blue_patch = mpatches.Patch(color='blue', label='The predicted data')
plt.legend(handles=[red_patch, blue_patch])

plt.plot(XPredict,YPredict, 'b')
plt.plot(XPredict,YTrue,'r')
```
